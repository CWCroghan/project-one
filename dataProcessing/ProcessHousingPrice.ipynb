# Loading Zip & Track cross walk from HUD


```python
data = pd.ExcelFile("../Resources/ZIP_TRACT_122021.xlsx")
#Get the sheet names
data.sheet_names
```

```python
# Read in data from SQLT0006 sheet
zipdf = pd.read_excel(data, sheet_name = 'SQLT0006', header=0)

zipdf.head()
```

```python
zipdf.info()
```

### Reading in the 500 city data to limit tracts of interest

```python
# Bringing in 500 city data to generate unique list of cities
city_file_path = "../ProcessedData/500_City_cleaned.csv"
citydf = pd.read_csv(city_file_path)
citydf.head()
```

```python
citydf.info()
```

```python
FIPS_want = citydf['TractFIPS'].unique().tolist()
```

```python
newdf = zipdf[zipdf.tract.isin(FIPS_want)]
newdf.info()
```

## Combine Housing data with zip-track crosswalk limiting to those that match

```python
house = HPIdf.merge(newdf, right_on='zip', left_on = 'Five-Digit ZIP Code' , how='inner') 

house.head()
```

```python
house = house.drop(['zip','Annual Change (%)', 'HPI with 1990 base', 'HPI with 2000 base'], axis = 1)
house.info()
```

Outputting informtion to try to find the cause of missing rows.

```python
comp = house.merge(citydf, right_on='TractFIPS', left_on = 'tract' , how='outer') 
comp.head()
```

```python
pd.set_option('display.max_rows', None)
check = comp[['TractFIPS', 'tract','usps_zip_pref_state', 'usps_zip_pref_city','StateAbbr','PlaceName']]
print(check)
```

```python
# Saving Cleaned data into csv
output_file_path = "../ProcessedData/HousingPrice_checl.csv"
check.to_csv(output_file_path, index=False)
```

```python
final = house[['tract', 'HPI']]

final = final.sort_values(["tract", "HPI"],
               axis = 0, ascending = True,
               na_position = "first")
final = final.groupby(final.tract).last()
final = final.reset_index()
final.info()
```

```python
# Saving Cleaned data into csv
output_file_path = "../ProcessedData/FinalHousing.csv"
final.to_csv(output_file_path, index=False)
```
