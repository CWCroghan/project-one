# Extract Data from Superfund  

```python
import pandas as pd
# Import the Numpy dependency
import numpy as np
```

```python
#Read the data into a DataFrame
sf_data = pd.ExcelFile("../Resources/SF_CRSI_OLEM.xlsx")

#Get the sheet names
sf_data.sheet_names
```

```python
# Read in data from crowdfunding_info sheet
df = pd.read_excel(sf_data, sheet_name = 'SF_CRSI' )

df.head()
```

```python
df = df.drop(['Site_Name','Status','Region'], axis=1)
```

```python
df.info()
```

```python
df.describe()
```

Cumulative Resilience Screening Index - community resilience to natural hazards

```python
# Find duplicates entries
print(f"Duplicate entries: {df.duplicated().sum()}")
```

```python
# Bringing in 500 city data to generate unique list of cities
city_file_path = "../ProcessedData/500_City_cleaned.csv"
citydf = pd.read_csv(city_file_path)
citydf.head()
```

```python
citydf['city_state'] = citydf.PlaceName + "_" +  citydf.StateAbbr
```

```python
city_want = citydf.city_state.unique()
print(city_want)
```

```python
State_file_path = "../Resources/StateNameAbbr.csv"
STdf = pd.read_csv(State_file_path)
STdf.head()
```

```python
df = df.merge(STdf, how='left', on='State')
```

```python
df.info()
```

```python
df['cityState'] = df.City + "_" +  df.StateAbbr
df.head()
```

```python
df = df[df.cityState.isin(city_want)]
df.info()
```

```python
sf = df[['Site_EPA_ID','City','County_FIPS','County','State','StateAbbr','Longitude','Latitude']]
```

```python
# Saving Cleaned data into csv
output_file_path = "../ProcessedData/Superfund.csv"
sf.to_csv(output_file_path, index=False)
```

```python
resildf = df[['Site_EPA_ID','City','County_FIPS','County','State','StateAbbr','Longitude','Latitude', 'EXPOSURE_SCORE', 'RISK_SCORE', 'CRSI_SCORE']]
```

```python
# Saving Cleaned data into csv
output_file_path = "../ProcessedData/ResilienceScore.csv"
resildf.to_csv(output_file_path, index=False)
```
