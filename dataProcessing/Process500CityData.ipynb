```python
import pandas as pd
```

```python
# Read the data into a Pandas DataFrame
df = pd.read_csv('C:/Users/carry/BootCamp/project-one/Resources/500_Cities__Census_Tract-level_Data__GIS_Friendly_Format___2019_release.csv')
df.head()
```

```python
df.info()
```

```python
newdf = df.drop(['ACCESS2_Crude95CI','ARTHRITIS_CrudePrev','ARTHRITIS_Crude95CI','BINGE_CrudePrev','BINGE_Crude95CI','BPHIGH_Crude95CI','BPMED_CrudePrev','BPMED_Crude95CI','CANCER_Crude95CI','CASTHMA_Crude95CI','CHD_Crude95CI','CHECKUP_Crude95CI','CHOLSCREEN_CrudePrev','CHOLSCREEN_Crude95CI','COLON_SCREEN_CrudePrev','COLON_SCREEN_Crude95CI','COPD_CrudePrev','COPD_Crude95CI','COREM_CrudePrev','COREM_Crude95CI','COREW_CrudePrev','COREW_Crude95CI','CSMOKING_Crude95CI','DENTAL_CrudePrev','DENTAL_Crude95CI','DIABETES_CrudePrev','DIABETES_Crude95CI','HIGHCHOL_CrudePrev','HIGHCHOL_Crude95CI','KIDNEY_CrudePrev','KIDNEY_Crude95CI','LPA_CrudePrev','LPA_Crude95CI','MAMMOUSE_CrudePrev','MAMMOUSE_Crude95CI','MHLTH_Crude95CI','OBESITY_CrudePrev','OBESITY_Crude95CI','PAPTEST_CrudePrev','PAPTEST_Crude95CI','PHLTH_CrudePrev','PHLTH_Crude95CI','SLEEP_CrudePrev','SLEEP_Crude95CI','STROKE_CrudePrev','STROKE_Crude95CI','TEETHLOST_CrudePrev','TEETHLOST_Crude95CI'], axis=1)
newdf.head()
```

```python
# Splitting Geolocation into Lat & Long
newdf[["Latitude","Longitude"]] = newdf["Geolocation"].str.split(',', n=1, expand=True)
newdf.head()
```

```python
# Removing the parens
newdf["Latitude"] = newdf["Latitude"].str.replace('(','', regex=False)
newdf["Longitude"] = newdf["Longitude"].replace('\)','', regex=True)
newdf.head()
```

```python
newdf = newdf.drop('Geolocation', axis=1)
```

```python
newdf.info()
```

```python
ren = newdf.rename(columns={"ACCESS2_CrudePrev": "Insurance", "BPHIGH_CrudePrev": "HighBloodPressure","CANCER_CrudePrev":"Cancer","CASTHMA_CrudePrev": "Asthma","CHD_CrudePrev": "HeartDisease", "CHECKUP_CrudePrev": "AnnualCheckUps","CSMOKING_CrudePrev": "Smokes", "MHLTH_CrudePrev": "MentalHealthIssues"})
ren.info()
```

```python
ren['Population2010'] = pd.to_numeric(ren['Population2010'], errors='coerce')
ren['Latitude'] = pd.to_numeric(ren['Latitude'], errors='coerce')
ren['Longitude'] = pd.to_numeric(ren['Longitude'], errors='coerce')

ren.info()
```

```python
# Drop null rows
ren = ren.dropna()
```

```python
ren.isnull().sum()
```

```python
# Find duplicates entries

print(f"Duplicate entries: {ren.duplicated().sum()}")
```

```python
# Saving Cleaned data into csv
output_file_path = "../ProcessedData/500_City_cleaned.csv"
ren.to_csv(output_file_path, index=False)
```
