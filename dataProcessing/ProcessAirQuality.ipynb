# Extract Data from Air Quaility by monitor
   

```python
import pandas as pd
import numpy as np
```

```python
# Read the data into a Pandas DataFrame
dfbase = pd.read_csv('../Resources/annual_conc_by_monitor_2017.csv')
dfbase.head()
```

```python
dfbase.columns
```

```python
dfbase['Parameter Name'].unique()
```

```python
pd.set_option('display.max_rows', None)
dfbase.groupby('Parameter Name')['1st Max Value'].describe()
```

```python
newdf = dfbase.drop(['State Code', 'County Code','POC', 'Datum','Method Name', 
                 'Year', 'Event Type', 'Observation Count','Observation Percent', 
                 'Completeness Indicator', 'Valid Day Count','Required Day Count',
                'Null Data Count','Secondary Exceedance Count',
                 'Certification Indicator', 'Num Obs Below MDL', 'Arithmetic Mean',
                 'Arithmetic Standard Dev', '1st Max DateTime', '2nd Max Value',  
                 '2nd Max DateTime', '3rd Max Value','3rd Max DateTime', 
                 '4th Max Value', '4th Max DateTime','1st Max Non Overlapping Value', 
                 '1st NO Max DateTime','2nd Max Non Overlapping Value', 
                 '2nd NO Max DateTime','99th Percentile', '98th Percentile', 
                 '95th Percentile','90th Percentile', '75th Percentile', 
                 '50th Percentile','10th Percentile','CBSA Name', 'Date of Last Change'], axis=1)
newdf.head()
```

```python
standards = ['Ozone 8-hour 2015','PM25 24-hour 2012']
limitdf = newdf[newdf['Pollutant Standard'].isin(standards)]
limitdf.info()
```

```python
limitdf.head()
```

# Creating Ozone variables

```python
ozone = limitdf[limitdf['Parameter Name'] == 'Ozone']
ozone.describe()
```

```python
ozone.columns
```

```python
ozone = ozone.rename(columns={'Exceptional Data Count': 'Ozone_Exceptional',
                              'Primary Exceedance Count': 'Ozone_ExceedCount',
                              '1st Max Value': 'Ozone_Max'})
```

```python
ozone.head()
```

```python
ozone = ozone.drop(['Parameter Code', 'Latitude', 'Longitude', 'Parameter Name',
                    'Sample Duration', 'Pollutant Standard', 'Metric Used',
                    'Units of Measure','Local Site Name','Address', 'State Name', 
                    'County Name', 'City Name'], axis=1)
```

# Creating PM Variables


```python
pm = limitdf[limitdf['Parameter Name'] == 'PM2.5 - Local Conditions']
pm.describe()
```

```python
pm = pm.rename(columns={'Exceptional Data Count': 'PM2.5_Exceptional',
                         'Primary Exceedance Count': 'PM2.5_ExceedCount',
                         '1st Max Value': 'PM2.5_Max',
                         'State Name': 'State',
                         'County Name': 'County', 
                         'City Name': 'City'
                        })
pm.info()
```

```python
pm = pm.drop(['Parameter Name','Sample Duration', 'Pollutant Standard', 
               'Local Site Name','Address','Parameter Code','Units of Measure',
               'Metric Used'], axis=1)
```

# Combine Ozone and PM 2.5

```python
df = pm.merge(ozone, how='left', on='Site Num')
df.head()
```

## Adding StateAbbr and generating a matching variable for 500 city

```python
State_file_path = "../Resources/StateNameAbbr.csv"
STdf = pd.read_csv(State_file_path)
STdf.head()
```

```python
df = df.merge(STdf, how='left', on='State')
```

```python
df.count()
```

```python
df['city_state'] = df.City + "_" +  df.StateAbbr
```

```python
df.info()
```

# Bring in 500 city data to limit to cities of interest and locate closest monitor

```python
# Bringing in 500 city data to generate unique list of cities
city_file_path = "../ProcessedData/500_City_cleaned.csv"
citydf = pd.read_csv(city_file_path)
citydf.head()
```

```python
citydf = citydf[['StateAbbr','PlaceName','PlaceFIPS','TractFIPS','Place_TractID',
                 'CountyFIPS','Latitude','Longitude']]
```

```python
citydf['city_state'] = citydf.PlaceName + "_" +  citydf.StateAbbr
```

```python
city_want = citydf.city_state.unique()
```

```python
df = df[df.city_state.isin(city_want)]
df.count()
```

```python
df.describe()
```

```python
# Saving Cleaned data into csv
output_file_path = "../ProcessedData/AirQuality_cleaned.csv"
df.to_csv(output_file_path, index=False)
```
