# Extract Data from Chronic Conditions Spending

```python
import pandas as pd
import numpy as np
```

```python
#Read the data into a DataFrame
sf_data = pd.ExcelFile("../Resources/County_Table_Chronic_Conditions_Spending_2017_cwc.xlsx")

#Get the sheet names
sf_data.sheet_names
```

```python
# Read in data from crowdfunding_info sheet
CCSdf = pd.read_excel(sf_data, sheet_name = 'Standardized Spending', header=4)

CCSdf.head()
```

```python
CCSdf.columns
```

```python
df = CCSdf.rename(columns={'State/County FIPS Code': 'FIPS',
    'Alcohol Abuse': 'AlcoholAbuse',
    "Alzheimer's Disease/Dementia ": 'Dementia',
    'Arthritis ': 'Arthritis',
    'Asthma ': 'Asthma',
    'Atrial Fibrillation ': 'AFHeartDisease',
    'Autism Spectrum Disorders ': 'Autism',
    'Cancer ': 'Cancer',
    'Chronic Kidney Disease ':'ChronicKidneyDisease', 
    'COPD ':  'COPD',
    'Depression ':  'Depression',
    'Diabetes ': 'Diabetes',
    'Drug Abuse/Substance Abuse': 'SubstanceAbuse',
    'HIV/AIDS': 'AIDS',
    'Heart Failure ': 'HeartFailure',
    'Hepatitis                                 (Chronic Viral B & C)': 'Hepatitis',
    'Hyperlipidemia ': 'Hyperlipidemia',
    'Hypertension ': 'Hypertension',
    'Ischemic Heart Disease ': 'CoronaryArtery',
    'Osteoporosis ': 'Osteoporosis',
    'Schizophrenia/Other Psychotic Disorders ': 'Psychotic',
    'Stroke ': 'Stroke',})
```

```python
df.columns
```

```python
cols = df.columns.to_series().reset_index(drop = True)
print(cols)
```

```python
cols = cols[2:]
print(cols)
```

```python
for c in cols:
    df[c] = pd.to_numeric(df[c], errors='coerce')
   
df.info()
```

```python
col_list = list(df)
print(col_list)
```

```python
col_list=col_list[3::]
print(col_list)
```

```python
df['TotalCost'] = df[col_list].sum(axis=1)
df.head()
```

```python
df["FIPS"].count()
```

```python
df['FIPS'] = df['FIPS'].astype(int)
display(df.dtypes)
```

# Reading in the 500 city data to limit counties of interest


```python
# Bringing in 500 city data to generate unique list of cities
city_file_path = "../ProcessedData/500_City_cleaned.csv"
citydf = pd.read_csv(city_file_path)
citydf.head()
```

```python
citydf = citydf[['StateAbbr','PlaceName','PlaceFIPS','TractFIPS','Place_TractID','CountyFIPS']]
```

```python
FIPS_want = citydf['CountyFIPS'].unique().tolist()

print(FIPS_want[0:5])
len(FIPS_want)
```

```python
newdf = df[df.FIPS.isin(FIPS_want)]
newdf.info()
```

```python
FIPS = df.FIPS.unique()
print(FIPS[0:5])
```

```python
FIPS2 = newdf.FIPS.unique()
print(FIPS2[0:5])
```

```python
linkeddf = citydf.merge(newdf, how='left', left_on='CountyFIPS', right_on='FIPS')
linkeddf.head()
```

```python
linkeddf.columns
```

```python
linkeddf = linkeddf[['State','StateAbbr','County', 'FIPS','Asthma','AFHeartDisease','COPD','Depression','HeartFailure','Hypertension','CoronaryArtery','Psychotic','Stroke','TotalCost']]
```

```python
linkeddf.head()
```

```python
df = linkeddf.drop_duplicates(ignore_index = True)
df = df.rename(columns={'FIPS': 'CountyFIPS'})
```

```python
df.count()
```

```python
# Saving Cleaned data into csv
output_file_path = "../ProcessedData/Chronic_Conditions_Spending.csv"
df.to_csv(output_file_path, index=False)
```
